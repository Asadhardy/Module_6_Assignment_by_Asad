```python
'''
Q10. Explain the concepts of vectorization and broadcasting in NumPy. How do they contribute to efficient array
operations?
Answers:
    1.Vectorization in Numpy:
        vectorization refers to the process of applying operations to entire arrays without using 
        explicit loops.
        It leverages low-level optimizations in the numpy library to make array opearions faster
        and more efficient.
        
        Instead of iterating through each element manually , numpy allows you to perform mathematical
        operations on whole array at once.
        
        Vectorization is implementing under the hood in numpy using highly optimized C and fortran libraries, 
        which is significantly speeds up computation
    Example
'''
# Withour vectorinization
arr = [1,2,3,4,5]
result = []
for i in arr:
    result.append(i*2)
print("Without Vectorization",result)

# With vectorization
arr_1 = [1, 2, 3, 4, 5]
vecto = np.array(arr_1) * 2
print("With Vectorization",vecto)
'''
Benefits:
    Vectorization operations are not only easier to write and read but also much faster because they avoid python loops
    and directly utilize optimized numpy methods.

    BroadCasting in Numpy:
        Broadcasting allows numpy to perform operations between arrays of different shapes in a way that they
        are "stretched" or "broadcasted" to a compatible shape.
        It avoids the need for manual reshaping or replication of smaller arrays, saving memory and computational 
        loads(operations).
Example:
'''
array = np.array([1,2,3])
scalar = 5
result = array + scalar
print("Broadcasted aarrya",result)
'''
In this example, the 5 is broadcasted to match the shape (3, ) so that element-wise addition can occur.

Broadcasting Rules:
    1.If two arrays have different numbers of dimension, Numpy pads the shape of the smaller array with ones on its
    left side.
    2.numpy compares shaped element-wise starting from the right, if dimensions are equal or one of them is 1,
    the array are compatible.
    3. If the array are compatible, they are virtually expanded (broadcasted) to the larger array's shape for the 
    operation.
Example:(Broadcasting with multi dimensional arrays)
'''
a = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
b = np.array([1, 0, 1,0])
result = a + b
print("Broadcasted result\n",result)
'''
In this example: b is broadcasted to match the shape (2, 4) before element wise addition is performed.

Contribution to Efficient Array Operations:
    1.Speed: both vectorization and broadcasting eliminates the need for explicit loops,
    making the operations faster as they leverage low-level optimizations.
    2.memory efficiency: broadcasting avoids the need to create large temporary arrays by virtually expanding smaller
    manner, which makes code easier to maintain and understand.
    
Insights:
    Together, vectorization and broadcasting form the backbone of efficient array operations in numpy.
    They allow for concise, high performance code, making Numpy an essential tool for numerical computing.
    These features optimize both computation time and memory usage, which is particularly beneficial for handling 
    large datasets in machine learning and scientific applications.
'''
print()
```
