```python
'''
Q9. Generate a NumPy array of 100 random integers between 0 and 1000. Find and display all prime numbers in
this array.
Answer:
    To generate a numpy array of 100 random integers between 0 to 1000 then find and display all prime numbers
    in that array,
    code are as following performing this action( to check if the number is prime and print it in the console)
Example Code:
'''
# Generating 100 random number from 0 to 1000
random_numbers = np.random.randint(0, 1001, size= 100)

# Displaying the Random integers generated
print("Random integers are: \n", random_numbers)

# Using a function to check from the generated random number which are the prime numbers
def is_prime(num):
    if num <=1:
        return False
    for i in range(2, int(num**0.5) +1):
        if num % i == 0:
            return False
    return True

# Using list comprihension for checking the prime number using creating a loop through function
checking_prime = [num for num in random_numbers if is_prime(num)]

# Displaying the Prime Numbers
print("Random generated Prime numbers are:\n" ,[int(num) for num in checking_prime]) # using loop for readibility of the code
'''
Explanation:
    1. np.random.randint(0, 1001, size = 100) , this generated an array of 100 integers in the range
        from 0 to 1000.
    2.The is_prime(n) function checks if n is a prime number by testing divisibility from 2 up to the square root of n.
    3. The i use a list comprehension to filter out prime prime numbers from the random integers using the 
        is_prime function

'''
print()
```

```python
'''
10. Create a NumPy array representing daily temperatures for a month. Calculate and display the weekly
averages.?
'''
import numpy as np

# Random daily temperatures for 30 days
daily_temperatures = np.random.randint(20, 40, size=30)  # Assuming temperatures between 20°C and 40°C
print("Daily Temperatures for the Month:\n", daily_temperatures)

# First 28 days reshape for 4 full weeks
weekly_temperatures = daily_temperatures[:28].reshape(4, 7)

# Calculate weekly averages for the first 4 weeks
weekly_averages = weekly_temperatures.mean(axis=1)

# Handling the last 2 days 
last_week_average = daily_temperatures[28:].mean()

# Display the weekly averages
for i, avg in enumerate(weekly_averages, 1):
    print(f"Week {i} average temperature: {avg:.2f}°C")

# Displaying the average for the last few days
print(f"Last 2 days average temperature: {last_week_average:.2f}°C")
'''
Explanation:
    1.Daily Temperatures: Generates random temperatures for 30 days.
    2.Weekly Calculation: The first 28 days are reshaped into 4 weeks, and their averages are calculated.
    3.Last 2 Days: The average for the last 2 days is calculated separately.
Insights:
    by using Numpy , we can efficiently work with arrays representing large data sets like daily temperatures,
    Reshaping and calculating weekly averages allows for organised analysis of data over specific time perious.
    this approach is highly efficient and scalable for similar tasks involving time series data.
'''
print()
```
